```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("train.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df['class'].value_counts()
```

```python
X = df.drop(['class'], axis=1)

y = df['class']
```

```python
from sklearn.model_selection import train_test_split

X_train1, X_test1, y_train1, y_test1 = train_test_split(X, y, test_size = 0.33, random_state = 42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf_gini = DecisionTreeClassifier(criterion='gini', max_depth=5, random_state=0)



clf_gini.fit(X_train1, y_train1)
```

```python
y_pred_gini = clf_gini.predict(X_test1)
y_pred_gini
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
```

```python
accuracy_score(y_test1,y_pred_gini)
```

```python
f1_score(y_test1,y_pred_gini)
```
